```python
import scipy.stats as stats
#Significance level=0.05
```

```python
#Crosstab and chi square test is carried out to find out whether there is any association between Stage and other qualitative  attributes.
```

```python
#Status
```

```python
a=pd.crosstab(df['Stage'],df['Status'] )
a
```

```python
pd.crosstab(df['Stage'],df['Status'] ).apply(lambda r: (r/r.sum())*100, axis=1)
```

```python
stats.chi2_contingency(a)
```

```python
#Since p<0.05 we conclude that there exist an association of Status with Stage
```

```python
#Drug
```

```python
b=pd.crosstab(df['Stage'],df['Drug'])
b
```

```python
pd.crosstab(df['Stage'],df['Drug']).apply(lambda r: (r/r.sum())*100, axis=1)
```

```python
stats.chi2_contingency(b)
```

```python
#Since p<0.05 we conclude that there exist an association of Drug with Stage
```

```python
#Handling missing values to do ANOVA
```

```python
n_df=df.copy()
```

```python
n_df.isna().sum()
```

```python
col_numeric = n_df.select_dtypes(include=(['int64', 'float64'])).columns
for col in col_numeric:
    n_df[col].fillna(n_df[col].median(), inplace=True)

col_obj = n_df.select_dtypes(include=(['object'])).columns
for col in col_obj:
    n_df[col].fillna(n_df[col].mode().values[0], inplace=True)

n_df.isna().sum()    
```
