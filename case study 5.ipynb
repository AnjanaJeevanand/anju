# Check whether all the zones generate the same amount of sales

Ho: The mean sales of different zones are same
    

H1: The mean sales of some of the zones may differ

```python
stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'], axis=0)
```

Conclusion:
Since P<0.05 We reject H0 and conclude that some of the zones differs in mean sales
That is all zones donot generate the same amount of sales
